* Consider a few bad examples:

#### Example 1: Tuskgee Syphilis Experiment: 

* Approx 600 rural afro american men in Tuskgee, Alabama thought they were receiving free healthcare 1932 - 1972. 
* In reality they were being cited regarding their natural progression of untreated syphilis.
* They were not treated, not even informed of treatment after penicillin invented.



#### Example 2: Milgram experiments:
* Yale professor Stanley Milgram in early 1960s
* Measured willingness of participants to obey an authority figure who instructed them to administer electric shocks to 3rd person
* 3rd person was an actor pretending to be shocked with gradually increasing responses and plea to stop
* participants subject to harm - (psychologivcal damage)


#### Example 3: Facebook:
* Cornell researcher: how emotions can spread
* news feed with positive/negative posts
* studied if posts from those were positive or negative
* Risk lower than other

#### Principles to consider:
1. <b>Risk:</b> What risks are participants being exposed to
<ul> 
* Main threshold: whether or not risk exceeds that of "minimal risk". <br>Minimal risk: Probability and Magnitude of harm that a participant would encounter in normal daily life
* Harm considered encompasses physical, psychologivcal and emotional, social and economic concerns.
* If the risk exceeds minimal risk, then informed consent is required.
</ul>
<br>
2. <b>Benefits:</b> What benefits might be outcome of study?
<br><br> 

3. <b>Alternatives:</b> What other choices do participants have?
<ul> 
* Eg. when testing out changes to a search enginer, participants always have the choice to use another search engine.
* If choices are fewer, it may be a case of coercion where participants don;t really have a choice in whether to participate or not, and how that balances agains the risk and benefits.
</ul> 


4. <b>Data Sensitivity:</b> What expectation of privacy and confidentiality do participants have?
<ul> 
* Do participants understand what data is being collected about them?
* How sensitive is the data? Does it include financial and health data?
* What harm would befall them should that data be made public?
* Can the data be tied to the individual, i.e. considered personally identifiable?
* Would they expect that data to be considered private and confidential?
* Is the data secure?
</ul> 
    
<hr>

#### Questions boil down to:

1. Are participants facing more than minimal risk?
2. Are users being informed?
3. What user identifiers are tied to the data?
4. What type of data is being collected?
5. What level of conidentiality and security is the data sebject to?

Informed consent should adress these questions
