1. Choose the Subject: Decide how to assign events to control or the experiment

    - Could be random, but for user visible change, a user may see version A, refresh and then see version B. 
      So for user visible change we want to assign people to those groups instead of events.
      
    - How to determine a person? 
        Each person might have multiple usernames
        Cookie: assigned to particular browser or device, but changing browsers/devices changes the cookie.



Unit of diversion: how we define what an individual subject. Commonly used include
    - User ID
        - Stable or unchanging across browsers/devices
        - Personally identifiable
    
    - Anonymous ID (Cookie)
        - Changes when you switch the browser/device
        - Users can clear cookies
    
    - Event
        - On every event, redecide if event in experiemnt/control
        - User may not be a consistent experience, only an option for non-visible changes
        - Eg: order change in ranked list
        
        
     - Device ID (less common - only available on mobile)
         - tied to specific device
         - unchangeable by user
        
     - IP Address
         - changes when user location changes

How to use unit of diversion? Three main considerations:

1. User consistency: 
   
   - If user_id used, users gets consistent experience across devices as long as they are signed in.
   
   - If testing a change that crosses the sign-in, sign-out border then user-ID doesn't work well.(eg. page layout, location of sign in bar) use cookie instead to get consistency across sign-in sign-out border but not across devices.
   
   - User visibility is one consideration to decide whether user_id or cookie is suitable. Other consideration is what you need to measure.
   
   - To measure learning effect, whether users adapting to change or latency change,  need stateful metric like user_id or cookie.
   
   
   

   
2. Ethical Considerations:

    - if data is identified, security and confidentiality questions must be answered
    - Informed consent needed if user identifiable.
     
    
3. Variability:
    - Unit of analysis is the denominbator of your metric. Eg. CTR = Clicks/Pageviews. Pageviews is the unit of analysis.
    
    - If unit of diversion is also a pageview, analytical variability likely close to empirical variability.
 
    - If unit of diversion is user_id or cookie, variablility will be much higher. Move to empirically given your unit of diversion.
    

- Measuring variability analytically makes an assumption about 
    - the fundamental distribution of the data,
    - what's considered to be independant (valid in event based diversion, not in cookie or user based diversion)

    
    
    

Other considerations when choosing a population:
    
1. Cookie/device diversion: proxies for users.
    - Users in A and B groups are mutually exclusive.
    - Event based diversion could have mix of people on both sides.
    - Intra-user experiement: expose a user to feature being on and off over time and check their behaviour over time windows.
    - Inter-leaved experiement:  expose user to A and B at the same time.



2. Who are you targeting?
    - Divisions of user space by browser/language/geo location.
    - How long they've been using website
    - Demographics.
        For large changes, (eg. international changes) these changes are recommended to be done upfront.

        For a change that is significant but we might want to restrict releaseing information to the public, a smaller group to test on is advisable.
        
    - Restrict to modern browsers?
    - Avoid diluting effect of experiment across a global population. If change only affects english speakers, only run in english speaking countries.
    

3. When to not target experiment in advance?
    - cannot identify which users will be affected by change
    - not sure it change works so need to target global population
    - not a consideration if the change affects 90% of traffic

4. Before launching:
    - Double check with Engineering team about the browsers etc the change affects
    - Ensure the test does not interfere with other currently run tests
    - same filters on targeted and un-targeted populations
    - Check for unintended effects on populations not targeted


Population vs Cohort

- Within population: people who enter experiement on both sides at the same time. 

- If groups of users have not started at the same time, other changes to website made in the time when only part of users were using the site, may affect the results

- Cohorts used when we want user stability to measure real effect on behavious relative to their history.
eg: when there is a learning effect, or when measuring something
    like increased usage of the site/mobile device
    
    

#### Sizing

While practical and statistical significance and sensitivity govern the size, following must also be considered in addition:
    - choice of metric, 
    - unit of diversion and 
    - population 

since these can affect variability of metric.

#### Sizing Triggering:

We might not always know what may be effective beforehand when it comes to sizing. How many users affected by change is sometimes a function of

    - which browsers will benefit?
    - language classification works when # of search results in that language are higher than number in some other language
    - keep track of who is already exposed to change

#### Duration 

- What is duration of exp
- when do I want to run exp? back to school? holidays good?
- What fraction of traffic to send through experiment?<br>
  OR what % of cookies sending through A and B?<br>
     
    Eg. 1Mn cookies needed. 
          if 100,000 users/day and 
          50% split in A and B required 
          
          then we need to run experiment for 10 days.
    
#### Exposure

expose only a few users with feature since
- with new UI feature we might not be sure how it works/users react.
- If it is a new new feature you are not sure about, we wouldn't want it to end up in the press.

- Run across multiple days to account for sources of variability with weekday vs weekend, holiday vs non-holiday
- if multiple tests on same feature run, try to run at the same time on smaller % of traffic to so all affectred by things like holidays and we have relative ordering against each other
